# Ejercicio final

Escoge un dataset de regresión o clasificación (el que tu quieras) y diseña una red neuronal para clasificarlo. El algoritmo de entrenamiento de incluir early stopping y algún método de regularización (dropout, $L_1$ o $L_2$).

Si se usa regresión, recomiendo usar el MSE como función de pérdida (https://docs.pytorch.org/docs/stable/generated/torch.nn.MSELoss.html).

Ejemplos clasificación:
- https://archive.ics.uci.edu/dataset/109/wine
- https://archive.ics.uci.edu/dataset/17/breast+cancer+wisconsin+diagnostic

Ejemplos regresión:
- https://www.kaggle.com/datasets/camnugent/california-housing-prices
- https://www.kaggle.com/code/satoru90/boston-housing-dataset-regression
- https://scikit-learn.org/stable/modules/generated/sklearn.datasets.load_diabetes.html
